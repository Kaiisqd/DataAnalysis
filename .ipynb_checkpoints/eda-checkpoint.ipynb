```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.subplots(figsize=(15,15))

plt.subplot(3,3,1)
plt.hist(data['in-store'])
plt.title('In-Store')

plt.subplot(3,3,2)
plt.hist(data['age'])
plt.title('Age')

plt.subplot(3,3,3)
plt.hist(data['items'])
plt.title('Items')

plt.subplot(3,3,4)
plt.hist(data['amount'])
plt.title('Amount')

plt.subplot(3,3,5)
plt.hist(data['region'])
plt.title('Region')
```
